```python
import datetime
import schedule
import urllib.request as req
from bs4 import BeautifulSoup

def scrapWorking():
  url = 'https://vip.mk.co.kr/newSt/price/price.php'
  data = req.urlopen(url)

  bsObj = BeautifulSoup(data, 'html.parser')
  #print(bsObj)
  price = bsObj.select_one('span#lastTick[6] > font.f3_b').string
  #print('ì‚¼ì„±ì „ì ì£¼ê°€ :',price)

  t = datetime.datetime.now()
  #print(t)
  fname = 'samsung/' + t.strftime('%Y-%m-%d %H:%M:%S') + '.txt'
  #print(fname)

  with open(fname, 'w') as f:
    f.write(price)

scrapWorking()  #1íšŒìš©

#ìŠ¤ì¼€ì¤„ëŸ¬ ì‚¬ìš©
schedule.every(5).seconds.do(scrapWorking)          #5ì´ˆì— í•œ ë²ˆì”©
#schedule.every(5).minutes.do(scrapWorking)         #5ë¶„ì— í•œ ë²ˆì”©
#schedule.every(5).hours.do(scrapWorking)           #5ì‹œê°„ì— í•œ ë²ˆì”©
#schedule.every().day.at('10:20').do(scrapWorking)  #í•˜ë£¨ 10ì‹œ 20ë¶„ì— í•œ ë²ˆì”©
#schedule.every().monday.do(scrapWorking)           #ë§¤ì£¼ ì›”ìš”ì¼

#ë¬´í•œë£¨í”„ë¡œ ì‘ë™ì‹œí‚¤ê¸°
#while True:
#  schedule.run_pending()
#  time.sleep(1) 
```

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

url = 'https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
datas = urllib.request.urlopen(url).read()
#print(datas)

bsObj = BeautifulSoup(datas, 'html.parser')
#print(bsObj)

title = bsObj.find('title').string
#print(title)  #ê¸°ìƒì²­ ìœ¡ìƒ ì¤‘ê¸°ì˜ˆë³´

wf = bsObj.find('wf')
#print(wf)

#ë„ì‹œ
city = bsObj.findAll('city')
#print(city)
cityDatas = []
for c in city:
  cityDatas.append(c.string)
#ì»¬ëŸ¼ìœ¼ë¡œ ì €ì¥
df = pd.DataFrame()
df['city'] = cityDatas
print(df.head())
print()

#ìµœì €ê¸°ì˜¨
tempMin = bsObj.select('location > province + city + data > tmn')  # + ëŠ” ìˆœë°©í–¥, - ëŠ” ì—­ë°©í–¥. province + city + dataëŠ” í˜•ì œ
#print(tempMin)
tempDatas = []
for t in tempMin:
  tempDatas.append(t.string)
  
#ì»¬ëŸ¼ìœ¼ë¡œ ì €ì¥
df['temp_min'] = tempDatas
print(df.head())
print()

#ì»¬ëŸ¼ëª… ë³€ê²½
df.columns = ['ì§€ì—­','ìµœì €ê¸°ì˜¨']
print(df.head())

#dfë¥¼ íŒŒì¼ë¡œ ì €ì¥
df.to_csv('ë‚ ì”¨ì˜ˆë³´.csv', index=False)
df.to_html('ë‚ ì”¨ì˜ˆë³´.html', index=False)
```

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import locale
import re

url = 'https://www.bbq.co.kr/menu/menuList.asp?cidx=114&cname=%ED%99%A9%EC%98%AC+%EC%8B%9C%EA%B7%B8%EB%8B%88%EC%B2%98'
datas = urllib.request.urlopen(url).read()
#print(datas)

bsObj = BeautifulSoup(datas, 'html.parser')
#print(bsObj)

#ë©”ë‰´
menu = bsObj.select('div.info > p.name')
#print(menu)
menuList = []
for m in menu:
  menuList.append(m.string)
#print(menuList)

#ê°€ê²©
pay = bsObj.select('p.pay')
#print(pay)

#'18,000ì›' ì½¤ë§ˆê°€ ë“¤ì–´ê°„ ë¬¸ìì—´ì„ ìˆ«ìë¡œ ë³€í™˜í•˜ê¸°
payList = []
locale.setlocale(locale.LC_ALL, 'en_US.UTF8')
for p in pay:
  #ë°©ë²•1)locale ëª¨ë“ˆ ì‚¬ìš©
  p_int = locale.atoi(p.string[:-1])

  #ë°©ë²•2)ì •ê·œí‘œí˜„ì‹ ì‚¬ìš©
  #p_int = int(re.sub(r'[^0-9\s]','',p.string))

  #ë°©ë²•3)replace í•¨ìˆ˜ ì‚¬ìš©
  #p_int = int(p.string.replace(',','')[:-1])

  payList.append(p_int)
#print(payList)
#print(type(payList[0]))

df = pd.DataFrame(payList, menuList).reset_index()
df.columns = ['menu','price']
print(df)
print()

print('ê°€ê²©í‰ê· :', round(df['price'].mean(), 2))
print('ê°€ê²©í‘œì¤€í¸ì°¨:', df['price'].std())
print('ìµœê³ ê°€:', max(df['price']))
print('ìµœì €ê°€:', min(df['price']))
print()

print(df[df['price']==max(df['price'])])
print(df[df['price']==min(df['price'])])
```

```python
import urllib.request
import pandas as pd

url = 'http://openapi.seoul.go.kr:8088/sample/json/SeoulLibraryTime/1/5/'
plainText = urllib.request.urlopen(url).read().decode()
#ìœ„ì˜ encoding, decodingê³¼ ë‹¤ë¥´ë‹¤. ì›¹ìƒì—ì„œ ì§ë ¬í™”ì‹œí‚¨ ë°ì´í„°ë¥¼ í’€ì–´ì£¼ëŠ” ì‘ì—…ì´ë‹¤.
print(plainText)
print(type(plainText))  #<class 'str'> ë¬¸ìì—´ -> ì‘ì—…ì„ ìœ„í•´ dict í˜•ì‹ìœ¼ë¡œ ë°”ê¿”ì£¼ì–´ì•¼ í•œë‹¤.
print()

jsonData = json.loads(plainText)
print(jsonData)
print(type(jsonData))   #<class 'dict'>
print()

libData = jsonData.get('SeoulLibraryTime').get('row')
print(libData)

name = libData[0].get('LBRRY_NAME')
print(name)
print()

#DataFrame ì‘ì„±
datas = []
for ele in libData:
  name = ele.get('LBRRY_NAME')
  tel = ele.get('TEL_NO')
  addr = ele.get('ADRES')
  #print(name + ' ' + tel + ' ' + addr)
  imsi = [name, tel, addr]
  datas.append(imsi)
print(datas)

libraryDf = pd.DataFrame(datas)
print(libraryDf)
```
