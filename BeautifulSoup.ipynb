```python
import datetime
import schedule
import urllib.request as req
from bs4 import BeautifulSoup

def scrapWorking():
  url = 'https://vip.mk.co.kr/newSt/price/price.php'
  data = req.urlopen(url)

  bsObj = BeautifulSoup(data, 'html.parser')
  #print(bsObj)
  price = bsObj.select_one('span#lastTick[6] > font.f3_b').string
  #print('삼성전자 주가 :',price)

  t = datetime.datetime.now()
  #print(t)
  fname = 'samsung/' + t.strftime('%Y-%m-%d %H:%M:%S') + '.txt'
  #print(fname)

  with open(fname, 'w') as f:
    f.write(price)

scrapWorking()  #1회용

#스케줄러 사용
schedule.every(5).seconds.do(scrapWorking)          #5초에 한 번씩
#schedule.every(5).minutes.do(scrapWorking)         #5분에 한 번씩
#schedule.every(5).hours.do(scrapWorking)           #5시간에 한 번씩
#schedule.every().day.at('10:20').do(scrapWorking)  #하루 10시 20분에 한 번씩
#schedule.every().monday.do(scrapWorking)           #매주 월요일

#무한루프로 작동시키기
#while True:
#  schedule.run_pending()
#  time.sleep(1) 
```

# 📝 **구글의 검색기능 사용**

검색결과의 일부를 브라우저 택으로 각각 출력  


```python
import requests
from bs4 import BeautifulSoup
import webbrowser

try:
  url = 'https://www.google.com/search?q={0}'
  inputData = '파이썬'  
  searchWord = url.format(inputData)
  #print(searchWord)   #'https://www.google.com/search?q=파이썬'

  #작업을 하다 읽어들이는데 문제가 생긴다면 헤더를 다음과 같이 추가해보자. 
  datas = requests.get(searchWord, headers={ 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'})
  #print(datas)

  bsObj = BeautifulSoup(datas.text, 'html.parser')
  #print(bsObj)

  ##rso > div:nth-child(1) > div > div > div.yuRUbf
  linkData = bsObj.select('div > div.yuRUbf > a')
  #print(linkData)

  for link in linkData[:3]:
    #print(link)
    #print(type(link), type(str(link)))
    #print(str(link).find('https'), ' ', str(link).find('onmousedown'))  #링크가 https부터 시작해 onmousedown 바로 앞까지 있으므로 이를 이용해 위치를 찾는다
    url2 = str(link)[str(link).find('https'):str(link).find('onmousedown')-2]  #위치를 index로
    print(url2)

    webbrowser.open(url2)

except Exception as err:
  print('에러 : ', err)
```

## ⛅ **기상청 제공 날씨 정보 읽기**

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

url = 'https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
datas = urllib.request.urlopen(url).read()
#print(datas)

bsObj = BeautifulSoup(datas, 'html.parser')
#print(bsObj)

title = bsObj.find('title').string
#print(title)  #기상청 육상 중기예보

wf = bsObj.find('wf')
#print(wf)

#도시
city = bsObj.findAll('city')
#print(city)
cityDatas = []
for c in city:
  cityDatas.append(c.string)
#컬럼으로 저장
df = pd.DataFrame()
df['city'] = cityDatas
print(df.head())
print()

#최저기온
tempMin = bsObj.select('location > province + city + data > tmn')  # + 는 순방향, - 는 역방향. province + city + data는 형제
#print(tempMin)
tempDatas = []
for t in tempMin:
  tempDatas.append(t.string)
  
#컬럼으로 저장
df['temp_min'] = tempDatas
print(df.head())
print()

#컬럼명 변경
df.columns = ['지역','최저기온']
print(df.head())

#df를 파일로 저장
df.to_csv('날씨예보.csv', index=False)
df.to_html('날씨예보.html', index=False)
```


## 🍗 치킨브랜드 사이트에서 정보 읽기

가격평균(소수첫째자리까지), 표준편차 구하기  
최고가, 최저가 제품 구하기  



```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import locale
import re

url = 'https://www.bbq.co.kr/menu/menuList.asp?cidx=114&cname=%ED%99%A9%EC%98%AC+%EC%8B%9C%EA%B7%B8%EB%8B%88%EC%B2%98'
datas = urllib.request.urlopen(url).read()
#print(datas)

bsObj = BeautifulSoup(datas, 'html.parser')
#print(bsObj)

#메뉴
menu = bsObj.select('div.info > p.name')
#print(menu)
menuList = []
for m in menu:
  menuList.append(m.string)
#print(menuList)

#가격
pay = bsObj.select('p.pay')
#print(pay)

#'18,000원' 콤마가 들어간 문자열을 숫자로 변환하기
payList = []
locale.setlocale(locale.LC_ALL, 'en_US.UTF8')
for p in pay:
  #방법1)locale 모듈 사용
  p_int = locale.atoi(p.string[:-1])

  #방법2)정규표현식 사용
  #p_int = int(re.sub(r'[^0-9\s]','',p.string))

  #방법3)replace 함수 사용
  #p_int = int(p.string.replace(',','')[:-1])

  payList.append(p_int)
#print(payList)
#print(type(payList[0]))

df = pd.DataFrame(payList, menuList).reset_index()
df.columns = ['menu','price']
print(df)
print()

print('가격평균:', round(df['price'].mean(), 2))
print('가격표준편차:', df['price'].std())
print('최고가:', max(df['price']))
print('최저가:', min(df['price']))
print()

print(df[df['price']==max(df['price'])])
print(df[df['price']==min(df['price'])])
```

# 📝 **JSON 데이터 처리하기**

JSON : Lightweight한 데이터 표현방식. 배열만 이해하면 처리가 가능하다.  



```python
import json

dic = {'name':'신선해', 'age':22, 'score':['90','80','100']}
print(dic, ' ', type(dic))
print()

print('웹상에서 하는 encode, decode와 다르다!')

print('encoding: 파이썬 객체를 문자열로 변환')
str_val = json.dumps(dic)
print(str_val, type(str_val))
print(str_val[0:10])
#print(str_val['name'])  #error


print()
print('decoding: 문자열을 파이썬 객체로 변환')
json_val = json.loads(str_val)
print(json_val, type(json_val))
print(json_val['name'])
#print(json_val[0:10])
print(json_val['score'])
print(json_val.values())
print(json_val.get('name'))
```

## 📚 **강남구 도서관 정보 일부 읽기**



```python
import urllib.request
import pandas as pd

url = 'http://openapi.seoul.go.kr:8088/sample/json/SeoulLibraryTime/1/5/'
plainText = urllib.request.urlopen(url).read().decode()
#위의 encoding, decoding과 다르다. 웹상에서 직렬화시킨 데이터를 풀어주는 작업이다.
print(plainText)
print(type(plainText))  #<class 'str'> 문자열 -> 작업을 위해 dict 형식으로 바꿔주어야 한다.
print()

jsonData = json.loads(plainText)
print(jsonData)
print(type(jsonData))   #<class 'dict'>
print()

libData = jsonData.get('SeoulLibraryTime').get('row')
print(libData)

name = libData[0].get('LBRRY_NAME')
print(name)
print()

#DataFrame 작성
datas = []
for ele in libData:
  name = ele.get('LBRRY_NAME')
  tel = ele.get('TEL_NO')
  addr = ele.get('ADRES')
  #print(name + ' ' + tel + ' ' + addr)
  imsi = [name, tel, addr]
  datas.append(imsi)
print(datas)

libraryDf = pd.DataFrame(datas)
print(libraryDf)
```
